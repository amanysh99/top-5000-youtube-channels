# Data cleaning (Replace "--" to NaN)

```python
df.head(50)
```

```python
df=df.replace("--", np.nan, regex=True)
```

```python
df.head(50)
```

# **Check null values**

```python
df.isnull().sum()
```

```python
percentage_missing=df.isnull().sum() *100 /len(df)
```

# **Visualizing the missing**

```python
sns.heatmap(df.isnull())
```

# **Drop the missing values**

```python
df.dropna(axis=0, inplace=True)
```

```python
sns.heatmap(df.isnull())
```

# **Data cleaning (Rank col)**

```python
df.head()
```

```python
df['Rank']=df['Rank'].str.replace("st" ,"")
df['Rank']=df['Rank'].str.replace("nd" ,"")
df['Rank']=df['Rank'].str.replace("rd" ,"")
df['Rank']=df['Rank'].str.replace("th" ,"")
```

```python
df['Rank'].head()
```

```python
df.head(-200)
```

```python
df['Rank']=df['Rank'].str.replace(",", "")
```

```python
df.head(-200)
```

```python
df['Rank'].dtype
```

```python
df['Rank'].astype(int)
```

# **Data Cleaning (Video uploads & Subscribers)**

```python
df.head()
```

```python
df.dtypes
```

```python
df['Video Uploads']=df['Video Uploads'].astype('int')
```

```python
df['Subscribers']=df['Subscribers'].astype('int')
```

```python
df.dtypes
```

```python
df['Grade'].head()
```
